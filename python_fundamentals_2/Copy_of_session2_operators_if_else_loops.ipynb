## ðŸ”¹ Step 1: What is `print()`?

---

## ðŸ”¹ Step 2: What does `>>` mean?

`>>` is called the **right shift operator**.

ðŸ‘‰ **Meaning (in simple words):**
It shifts the **binary (0 and 1) form of a number to the right**.

Each shift to the right means:

* divide the number by **2**
* and **throw away the remainder**

---

## ðŸ”¹ Step 3: Convert `4` into binary

Decimal number:

```
4
```

Binary form of 4:

```
100
```

(You donâ€™t need to memorize this nowâ€”just follow)

---

## ðŸ”¹ Step 4: Shift right by `2`

```python
4 >> 2
```

Means:
ðŸ‘‰ move the bits **2 steps to the right**

```
Before: 100
After 1 shift: 010
After 2 shifts: 001
```

Now:

```
001 (binary) = 1 (decimal)
```

---

## ðŸ”¹ Step 5: Final Output

```python
print(4 >> 2)
```

ðŸ“¤ **Output:**

```
1
```

---

## ðŸ§  Super-Easy Shortcut (MOST IMPORTANT)

You can think of right shift like this:

```
number >> n  =  number Ã· (2â¿)
```

So:

```python

```

## ðŸ• Real-Life Example

So you end up with **1 pizza** ðŸ•
